```python
import numpy as np
import scipy.stats as stats #new library
import matplotlib.pyplot as plt
```

```python
# Step 1: Generate the Dataset
k = 55
np.random.seed(42)  # for reproducibility
data = 60 + np.random.randint(0, 15, size=12) + (k % 5)

# Step 2: State the Hypotheses
population_mean = 65  # hypothesized mean

# Step 3: Perform one-sample t-test
t_stat, p_value = stats.ttest_1samp(data, population_mean)
df = len(data) - 1
sample_mean = np.mean(data)

# Step 4: Compute 95% Confidence Interval
confidence = 0.95
alpha = 0.05
se = stats.sem(data)  # standard error of mean
ci = stats.t.interval(confidence, df, loc=sample_mean, scale=se)

# Step 5: Print Results
print("Sample Mean:", round(sample_mean, 2))
print("Hypothesized Mean:", population_mean)
print("t-statistic:", round(t_stat, 3))
print("Degrees of Freedom:", df)
print("p-value:", round(p_value, 4))
print("95% Confidence Interval:", (round(ci[0], 2), round(ci[1], 2)))
```

```python
# Decision
if p_value < alpha:
    print("Decision: Reject H0 (significant difference).")
else:
    print("Decision: Fail to reject H0 (no significant difference).")
```

```python
# Step 6: Visualization
# -----------------------------
plt.figure(figsize=(6,4))
plt.errorbar(1, sample_mean, yerr=se, fmt='o', capsize=5, label="Sample Mean Â±1 SE")
plt.axhline(population_mean, color='r', linestyle='--', label=f"Hypothesized Mean = {population_mean}")
plt.xticks([])
plt.ylabel("Exam Scores")
plt.title("Sample Mean vs Hypothesized Mean")
plt.legend()
plt.show()
```
