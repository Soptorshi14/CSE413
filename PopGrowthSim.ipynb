```python
import random
import matplotlib.pyplot as plt
```

```python
# Initial settings
initial_population = 1000000
years = 20

population = initial_population
populations = [population]
colors = []        # store color for each year bar
disaster_years = []  # store disaster year numbers

for year in range(1, years + 1):

    disaster = random.random() < 0.10  # 10% chance

    if disaster:
        growth_rate = random.uniform(-0.10, -0.02)
        colors.append("red")           # disaster year marked red
        disaster_years.append(year)
        event = "DISASTER YEAR"
    else:
        growth_rate = random.uniform(-0.01, 0.05)
        colors.append("blue")          # normal year marked blue
        event = "Normal year"

    population = int(population * (1 + growth_rate))
    populations.append(population)

    print(f"Year {year}: Population = {population}  Growth={growth_rate*100:.2f}%  --> {event}")
```

```python
# Plot bar chart
plt.figure(figsize=(12,6))
plt.bar(range(1, years + 1), populations[1:], color=colors)
plt.xlabel("Year")
plt.ylabel("Population")
plt.title("Population Growth Simulation (Red = Disaster Year)")
plt.xticks(range(1, years + 1))
plt.show()

print("\nDisaster years:", disaster_years)
```
